# Bar plot

## bar plt is represent category data with rectangular bars

```python
data = {'c': 20, 'Java': 30,'C++': 15, 'Python': 45}
```

```python
courses = list(data.keys())
courses
```

```python
values = list(data.values())
values
```

```python
courses = ['C', 'Java', 'C++', 'Python']
```

```python
values = [20, 30, 15, 45]
```

```python
plt.bar(courses,values)
```

```python
plt.bar(courses,values, color ='orange')
```

```python
plt.bar(courses,values, color = ['orange', 'red', 'blue', 'green'], edgecolor = ['black'])
```

```python
fig, ax = plt.subplots()
bars = ax.bar(courses, values)

ax.bar_label(bars)
plt.bar(courses, values)
```

```python
plt.bar(courses, values)
plt.xlabel('Cources offered')
plt.ylabel('Number of student enrolled')
plt.title('student enrolled in differnt courses')
plt.show()
```

# pie plot

```python
Fruits = ["Mango", "Banana", "Apple", "kiwi"]
Numbers = [10, 20, 12, 7]
```

```python
plt.pie(Numbers)
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits)
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits, autopct='%.2f')
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits, autopct='%.2f%%')
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits, autopct='%.2f%%', explode = [0, 0.1, 0, 0.2])
plt.show()
```
